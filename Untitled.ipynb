```python
%matplotlib inline

import matplotlib.pyplot as plt

def plot_waypoint_file(filename):
    
    f = open(filename,'r')
    x = []
    y = []
    for s in f:
        a,b = s.split(',')[0:2]
        x.append(a)
        y.append(b)
    f.close()
    plt.figure(figsize=(8,6), dpi=160)
    plt.plot(x,y,'x')
    
plot_waypoint_file('data\churchlot_with_cars.csv')
#plot_waypoint_file('data\sim_waypoints.csv')
plot_waypoint_file('data\wp_yaw_const.csv')
```

```python
# Test numpy vectorization for velocity profile in waypoint updater
import numpy as np
import math

stop_idx = 3
xyz = np.asarray([[1,2, 3],[1,2,4],[1,2,99],[1,2,103],[1,2,104],[1,5,104]])
print(xyz.shape)

# Compute the cumulative distance between points
cumulative_distances = np.cumsum(np.sqrt(np.sum(np.square(xyz[1:,:] - xyz[:-1,:]), axis = 1)))
print(cumulative_distances.shape)

# Compute distance to stopping point
stop_distances = np.maximum(0, cumulative_distances[stop_idx] - cumulative_distances)
print(stop_distances)

# Compute velocity needed to cause uniform deceleration
velocities = np.sqrt(2*5*stop_distances)
print(velocities)
```
